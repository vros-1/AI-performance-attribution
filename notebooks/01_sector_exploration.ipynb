```python
from pathlib import Path
import pandas as pd
data_path = Path ("..") / "data" / "sector_attribution_sample.csv"
df = pd.read_csv (data_path, parse_dates = ["date"])
df.head()
```

```python
df.info()
```

```python
df["sector"].vaue_counts()
```

```python
df.describe()
```

```python
sector_summary =(
    df.groupby("sector")[["allocation_effect", "selection_effect", "interaction_effect","excess_return"]]
        .sum()
        .sort_values ("excess_return", ascending= False)
)
sector_summary
```

```python
output_path = Path (". .") / "outputs" / "sector_attribution_for_tableau.csv"
sector_summary.reset_index().to_csv(output_path, index= False)
output_path
```
